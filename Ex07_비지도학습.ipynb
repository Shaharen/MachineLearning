## iris 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

### PCA
- 주성분 분석을 사용해서 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2) # 전체 데이터를 2차원으로 축소
```

```python
pca.fit(X)
```

```python
X_trans = pca.transform(X)
X_trans
```

```python
# 축소된 데이터 시각화
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```

```python
X = people['data']
y = people['target']
```

```python
# 픽셀값 범위를 0 ~ 1
X = X / 255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.3 ,
                                                   random_state = 10)
```

```python
knn.score(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
pca = PCA(n_components=100 , whiten = True)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
knn.fit(X_train_pca , y_train)
```

```python
knn.score(X_train_pca, y_train)
```

```python
knn.score(X_test_pca, y_test)
```

## 2차원 데이터셋

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X , y = make_blobs(random_state = 2)
```

```python
X.shape
```

```python
y
```

```python
# 군집 모델 생성
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
pred = kmeans.predict(X)
```

```python
pred
```

```python
# 중심 포인트의 좌표
kmeans.cluster_centers_
```

```python
# 생성한 데이터가 나눈 그룹
for n_cluster in range(3):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(3):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```

```python
# KMeans
## 모든 방향의 중요도가 같다고 생각ㄱ
## 데이터가 특정한 모양을 가지고 있는 경우 그룹을 잘 만들지 못한다
```

## 반달 데이터셋

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, random_state=0, noise=0.05)
```

```python
for n_cluster in range(2):
    data = X[y==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
pred = kmeans.predict(X)
```

```python
center = kmeans.cluster_centers_
for n_cluster in range(2):
    data = X[pred==n_cluster]
    plt.scatter(data[:, 0], data[: , 1], label = n_cluster)
    plt.plot(center[n_cluster , 0] , center[n_cluster , 1] , marker = '^' , c = 'black')

plt.legend()
plt.show()
```
